### Valores extremos (outliers)

```python
# FUNCIÓN QUE DEVUELVE LOS NÚMEROS DE FILA EN LOS QUE LA VARIABLE EN CUESTIÓN ES OULIER
def outliers(var):
  q1=var.quantile(0.25)
  q3=var.quantile(0.75)
  riq=q3-q1
  sup=q3+1.5*(riq)
  inf=q1-1.5*(riq)
  outl=(var>sup) | (var<inf)
  return outl
```

```python
# VEMOS CANTIDAD DE OUTLIERS DE CADA VARIABLE NUMÉRICA
for i in numericas:
    print(f"Cantidad de outliers de la variable **{i}**: [{outliers(df2[i]).sum()}]")
```

### Resumen numérico de cada variable (frecuencias). Usad la función que tenemos para generar las tablas de frecuencias

```python
def frequencia (d,c,dec=2):
  tabla = d[c].value_counts()
  tabla = pd.DataFrame(tabla)
  tabla["%"] = round(100*tabla[c]/len(d),dec)
  return tabla
```

```python
for i in categoricas:
  print("\n"+i.upper(),"\n")
  print(f"{frequencia(df,i)}")
```
